```python
from sympy import Symbol, symbols, factor, expand, pprint, simplify, solve, sin, cos
from sympy.plotting import plot
```

```python
x = Symbol('x')
x + x + 1
```

```python
x,y,z = symbols('x,y,z')
s = x*y + x*y
s
```

```python
# 展開
p = (x+2)*(x+3)
p
```

```python
expand(p)
```

```python
# 因数分解
expr = x**2 - y**2
factor(expr)
```

```python
# pretty print
expr = x**3 + 3*x**2*y + 3*x*y**2 + y**3
print(expr)
print('-----------')
pprint(expr)
```

```python
# 値の代入
expr = x**2 + 2*x*y + y**2
expr.subs({x:1, y:2})
```

```python
# x = 1 - y を代入
expr.subs({x:1-y})
```

```python
# 簡単化 (x = 1　- y のとき、x^2 + 2xy + y^2 = 1は恒等式)
simplify(expr.subs({x:1-y}))
```

```python
# 方程式を解く
# 二次方程式
expr = x**2 + 5*x + 4
solve(expr, dict=True)
```

```python
# 二次方程式(虚数解)
expr = x**2 + x + 1
solve(expr, dict=True)
```

```python
# 連立方程式
expr1 = 2*x + 3*y - 6
expr2 = 3*x + 2*y - 12
solve((expr1, expr2), dict=True)
```
